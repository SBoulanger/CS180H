Import needed libraries:

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import time
```

```python
cap = cv2.VideoCapture(2)
print("Say Cheese!")
ret,img = cap.read()
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
print(img)
plt.imshow(img)
```

```python
largethirds = np.zeros((img.shape[0],img.shape[1],img.shape[2]))
fimg = largethirds.copy()
smallthirds = np.full((3,3,3),1/3)
largethirds[0:3,0:3] = smallthirds
timeS = time.time()
for j in range(1,img.shape[0]-1):
    for i in range(1,img.shape[1]-1):
        temp = np.roll(np.roll(largethirds,i-1,-1),j-1,-2)
        fimg = fimg + np.multiply(temp,img)
timeE = time.time() - timeS
```

```python
print(timeE)
print(timeE/60)
plt.imshow(fimg)
```
